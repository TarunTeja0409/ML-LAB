```
import pandas as pd
titanic = pd.read_csv("titanic.csv")
print(titanic.isnull().sum())
```

```
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
titanic['Fare'].fillna(titanic['Fare'].mean(), inplace=True)
```

```
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
titanic['Fare'] = titanic['Fare'].fillna(titanic['Fare'].mean())
titanic['Embarked'] = titanic['Embarked'].fillna(titanic['Embarked'].mode()[0])
titanic['Cabin'] = titanic['Cabin'].fillna('Unknown')
```

```
titanic.to_csv('titanic_cleaned.csv', index=False)
print("Dataset Saved Successfully!")
```

```
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder for 'Sex' and 'Embarked'
le_sex = LabelEncoder()
le_embarked = LabelEncoder()

# Fit and transform 'Sex' column
titanic['Sex'] = le_sex.fit_transform(titanic['Sex'])

# Fit and transform 'Embarked' column
titanic['Embarked'] = le_embarked.fit_transform(titanic['Embarked'])

print("Unique values in 'Sex' after encoding:", titanic['Sex'].unique())
print("Unique values in 'Embarked' after encoding:", titanic['Embarked'].unique())
print(titanic.head())
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

### Subtask:
Reload the cleaned Titanic dataset, explicitly convert 'Sex' and 'Embarked' columns to string type, and then apply Label Encoding to ensure correct numerical representation and resolve the issue of all values being encoded as '0'.


**Reasoning**:
The previous Label Encoding resulted in all values being encoded as '0', indicating a potential issue with the column data types or how the encoder was applied. To resolve this, I need to reload the cleaned dataset, explicitly convert the 'Sex' and 'Embarked' columns to string type to ensure they are treated as categorical strings before encoding, and then apply the LabelEncoder to get proper numerical representations. This will address the problem of `Sex` and `Embarked` having all values as `NaN` in the `X` DataFrame, which caused the `ValueError` during model training.



```
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# 1. Reload the cleaned Titanic dataset
titanic = pd.read_csv('titanic_cleaned.csv')

# 2. Explicitly convert 'Sex' and 'Embarked' columns to string type
titanic['Sex'] = titanic['Sex'].astype(str)
titanic['Embarked'] = titanic['Embarked'].astype(str)

# Initialize LabelEncoder for 'Sex' and 'Embarked'
le_sex = LabelEncoder()
le_embarked = LabelEncoder()

# 3. Fit and transform 'Sex' column
titanic['Sex'] = le_sex.fit_transform(titanic['Sex'])

# 4. Fit and transform 'Embarked' column
titanic['Embarked'] = le_embarked.fit_transform(titanic['Embarked'])

# 5. Print unique values in 'Sex' and 'Embarked' after encoding
print("Unique values in 'Sex' after encoding:", titanic['Sex'].unique())
print("Unique values in 'Embarked' after encoding:", titanic['Embarked'].unique())

# 6. Print the head of the DataFrame to observe changes
print(titanic.head())
```

**Reasoning**:
Now that the 'Sex' and 'Embarked' columns have been correctly encoded, I need to re-run the train-test split using the updated `titanic` DataFrame to create `X_train`, `X_test`, `Y_train`, and `Y_test`. This will prepare the data for training the Logistic Regression model.



```
from sklearn.model_selection import train_test_split
X = titanic[['Age', 'Sex', 'Fare', 'Embarked']]
Y = titanic['Survived']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print("Training Data Size:", X_train.shape)
print("Testing Data Size:", X_test.shape)
```

**Reasoning**:
Now that the `Sex` and `Embarked` columns are correctly encoded and the data has been re-split, I will re-train the Logistic Regression model and evaluate its accuracy. This step confirms the fix for the previous `ValueError` and completes the primary task.



```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression(max_iter=1000) # Increased max_iter for convergence
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```
